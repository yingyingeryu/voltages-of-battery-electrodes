```python
import pandas as pd
import numpy as np
import matminer
from sklearn import preprocessing
from sklearn.model_selection import ShuffleSplit
```

```python
data = pd.read_csv('ML_capacity.csv')
```

```python
data
```

```python
data=data[data['working_ion']=='Li']
```

```python
df1=data[data['average_voltage']< 6.5]
df = df1[df1['average_voltage']> -1.5]
```

```python
y = df ['average_voltage'].values
excluded = ['Unnamed: 0','formula_charge', 'formula_discharge', 'structure',
            'average_voltage', 'working_ion', 'compound possible','volume_change','energy_vol','energy_grav',
           'capacity_grav','capacity_vol','discharge_frac','charge_frac','max_delta_volume']
X = df.drop(excluded, axis=1)
```

```python
scaler = preprocessing.MinMaxScaler().fit(X)
X_transformed = scaler.transform(X)
```

## GRCV_SVR

```python
# Set the parameters by cross-validation
tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-2, 0.1, 1, 10],
                     'C': [0.1, 1, 10, 100, 1e3], 'epsilon': [1e-3, 1e-2, 0.1, 1, 10]}]

scores = 'recall'
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svr = SVR (cache_size=200, coef0=0.0, degree=3)

cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=10)

GRCV_SVR = GridSearchCV(svr, tuned_parameters, cv=cv)
```

```python
GRCV_SVR.fit(X_transformed, y)
```
